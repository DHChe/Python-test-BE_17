[문제 1]
사용자로부터 이름과 나이를 입력 받아, "안녕하세요, [이름]님! 당신은 [나이]살입니다."라고 출력하는 함수 greet_user를 작성하세요.

```python
def greet_user(name, age):         
    print(f"안녕하세요, {name}님! 당신은 {age}살입니다.")

while True:
    name = input("당신의 이름을 입력하세요: ")
    if name == None or len(name) < 2:
        print("이름은 최소 2글자 이상 입력해야 합니다.")
        # continue (생략)
    else:
        break
        
while True:
    input_age = input("당신의 나이를 입력하세요: ") # input은 문자열을 반환함!!

    try:
        # 정수로 변환 시도
        age = int(input_age)
        # 정수 변환 성공했다면 나이 검증 로직 if문으로 이동, 아니라면 예외처리로 이동
        if age <= 0:
            print("나이는 음수일 수 없습니다.")
        else:
            break
        
    except ValueError:
        # int 변환 중 ValueError 발생하면
        print("나이는 숫자만 입력해 주세요.")

greet_user(name, age)

```

```python
MIN_NAME_LENGTH = 2
MIN_AGE = 1

def greet_user(name: str, age: int) -> None:
    """
    사용자에게 환경메세지를 출력 합니다.
    """
    print(f"안녕하세요, {name}님 당신은 {age}살 입니다.")

def get_valid_name() -> str:
    """
    사용자로부터 이름을 입력받아 유효성을 확인합니다.
    """
    while True:
        name = input("당신의 이름을 입력하세요: ").strip() # 양쪽 공백제거
        if len(name) < MIN_NAME_LENGTH:
            print(f"이름은 최소 {MIN_NAME_LENGTH} 글자 이상 입력해야 합니다.")
            # continue (생략)
        else:
            return name

def get_valid_age() -> int:
    """
    사용자로부터 나이를 입력받아 유효성을 확인합니다.
    """
    while True:
        input_age = input("당신의 나이를 입력하세요: ")
        try:
            age = int(input_age)
            if age >= MIN_AGE:
                return age
            print(f"나이는 {MIN_AGE}살 이상이어야 합니다.")
        except ValueError:
            print("숫자만 입력하세요.")

name = get_valid_name()
age = get_valid_age()

greet_user(name, age)


            
```

[문제 2] 여러 숫자를 입력받아 그 합계를 반환하는 함수 sum_numbers를 작성하세요. 이 때, 가변 매개변수를 사용하세요.

```python
# 1. 몇개 인지 모르는 숫자를 입력 받아야 하고,
# 2. 숫자만 입력 받아 합계를 구해야 한다.
# 3. 가변 매개변수(*args)를 사용해야 합니다.

def sum_numbers(*args: int) -> int:
    """
    사용자로부터 숫자를 입력받아 합계를 반환합니다.
    """
    result = 0
    for i in args:
        result += i
    return result

def user_input_numbers() -> list:
    """
    사용자로부터 입력받은 숫자의 유효성을 검증합니다.
    """
    input_numbers = []
    while True:
        input_num = input("숫자를 입력하세요: ").strip().lower() # 사용자가 q를 입력할때 대문자 또는 공백으로 인한 오류를 해결하기 위해
                
        if input_num == "q":
            break
        try:
            number = int(input_num)
            input_numbers.append(number)
        except ValueError:
            print("숫자 또는 'q'만 입력하세요.")

    return input_numbers

num_list = user_input_numbers()
print(f"입력하신 숫자는 {num_list}이고\n합계는 {sum_numbers(*num_list)}입니다.")


            





 
```

[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n: int) -> int:
    """
    0부터 n까지의 숫자 중 홀수만 생성하는 제네레이터 함수입니다.
    n은 함수를 사용할 때 결정합니다.
    """
    for i in range(1, n+1, 2):
        yield i
        

list_oddnum = odd_numbers(10)
print(type(list_oddnum))
print(list(list_oddnum))
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n):

    if n <= 1:
        return 1

    return n * factorial(n-1)

print(factorial(3))
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(str1, str2):

    if len(str1) > len(str2):
        return str1
    elif len(str1) == len(str2):
        return str1
    return str2

print(longer_string("banana", "onion"))
print(longer_string("cat", "elephant"))
print(longer_string("dog", "cat"))
```
