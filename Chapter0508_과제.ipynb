## 모험가들의 레벨업 도전

### [배경 이야기]

당신은 마법사, 전사, 도적, 힐러 네 명의 모험가를 이끄는 리더입니다. 

각 캐릭터는 힘, 민첩, 지능, 체력이라는 능력치를 가지고 있습니다. 이 능력치들을 이용해 다양한 도전을 해결하고 사냥을 통해 경험치를 얻어 레벨업을 해야 합니다.

### [동료 캐릭터와 능력치 정보]

__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다. 

마법을 사용하려면 지능이 50 이상이어야 합니다. 

마법사의 지능이 50 이상인지 확인하세요.

```python
# 캐릭터들의 스탯을 딕셔너리로 할당

mage = {
    'str': 10, 
    'dex': 20,
    'intel': 80,
    'vit': 50
}

warrior = {
    'str': 80, 
    'dex': 30,
    'intel': 20,
    'vit': 100
}

rog = {
    'str': 40, 
    'dex': 80,
    'intel': 30,
    'vit': 60
}

heal = {
    'str': 20, 
    'dex': 40,
    'intel': 70,
    'vit': 70
}

# 마법사(mage)의 지능 스탯이 50이상인지?
if mage['intel'] >= 50:
    print(f"지능이 50 이상이므로 마법을 사용할 수 있습니다.")
else:
    print(f"지능이 부족합니다. 수련하세요.")



```

<br>

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다. 

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
# 전사의 힘이 60이상, and 체력이 80이상을 모두 만족해야 
if warrior['str'] >= 60 and warrior['vit'] >= 80:
    print(f"전사의 힘이 60이상이고 체력이 80이상입니다. 강력한 공격을 사용하세요")
    
# 둘중 하나의 스탯이라도 모자라면     
elif warrior['str'] < 60 or warrior['vit'] < 80:
    print(f"강력한 공격을 사용할 수 없습니다. ")
```

<br>

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다. 

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
if rog['dex'] >= 70 or rog['str'] >= 50:
    print(f"몬스터를 회피하고 공격할 수 있습니다.")
elif rog['dex'] < 70 and rog['str'] < 50:
    print(f"공격을 회피할 수 없습니다.")
```

<br>

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다. 

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
if heal['intel'] >= 60 and heal['vit'] >= 50:
    print(f"힐 마법을 사용해 치유하세요.")
elif heal['intel'] < 60 or heal['vit'] < 50:
    print(f"힐 마법을 사용할 수 없습니다.")
```

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다. 


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
mage['xp'] = 120
warrior['xp'] = 150
rog['xp'] = 110
heal['xp'] = 90

characters = {
    'mage' : mage,
    'warrior' : warrior,
    'rog' : rog,
    'heal' : heal
}

for job, stats in characters.items():
    if stats['xp'] >= 100:
        print(f"{job} 캐릭터는 경험치가 {stats['xp']} 이상이므로 레벨업 가능합니다.")
    else:
        print(f"{job} 캐릭터는 경험치가 부족합니다. 현재 경험치 {stats['xp']}/100 ")
```
