[문제 1]

사용자로부터 두 개의 숫자를 입력받아 나눗셈을 수행하는 프로그램을 작성하세요. 

[예외처리]
- 입력 값이 숫자가 아닐 때 발생할 수 있는 ValueError
- 나누는 수가 0일 때 발생할 수 있는 ZeroDivisionError
- 위 2가지 예외처리 후 사용자에게 이해하기 쉬운 메시지 작성 후 출력되도록 코드를 작성

````python
def div_cal():

    while True:
        try:
            num1 = int(input("첫번째 숫자를 입력하세요: "))
            break
        except ValueError:
            print("유효한 숫자를 입력해주세요.")

    while True:
        try:
            num2 = int(input("두번째 숫자를 입력하세요: "))
            break
        except ValueError:
            print("유효한 숫자를 입력해주세요.")

    # 정상적으로 숫자 2개를 입력 받았을때
    try:
        result = num1 / num2
        print(f"결과: {result}")
    
    # 0으로 나누려고 할때 예외 처리
    except ZeroDivisionError:
        print("0으로는 나눌 수 없습니다.")

div_cal()


````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def timer_decorator(func):
    """함수 실행 시간을 측정하는 데코레이터"""
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"함수 실행 시간: {execution_time:.2f}초")
        return result
    return wrapper

# 테스트용 함수
@timer_decorator
def example_function():
    """테스트용 함수"""
    time.sleep(1)  # 1초 대기
    return "함수 실행 완료"

# 데코레이터 적용된 함수 실행
result = example_function()
print(result)
````

[문제 8]

여러분이 자주 사용하는 수학적 연산<br>
(예: 삼각형 넓이, 원의 넓이, 직육면체의 넓이)을 수행하는 함수들을 포함하는 mymath.py라는 이름의 모듈을 만드세요. 

이 모듈을 사용하여 작성된 함수들을 실제로 임포트하고 사용하는 간단한 프로그램을 작성하세요.

````python
# mymath.py 모듈 생성
import math

def triangle_area(base, height):
    """삼각형의 넓이를 계산합니다."""
    return 0.5 * base * height

def circle_area(radius):
    """원의 넓이를 계산합니다."""
    return math.pi * radius ** 2

def rectangular_prism_area(length, width, height):
    """직육면체의 표면적을 계산합니다."""
    return 2 * (length * width + width * height + height * length)

def rectangular_prism_volume(length, width, height):
    """직육면체의 부피를 계산합니다."""
    return length * width * height
````

````python
import mymath

# 삼각형 넓이 계산
triangle_base = 10
triangle_height = 5
triangle_result = mymath.triangle_area(triangle_base, triangle_height)
print(f"밑변 {triangle_base}, 높이 {triangle_height}인 삼각형의 넓이: {triangle_result}")

# 원의 넓이 계산
circle_radius = 7
circle_result = mymath.circle_area(circle_radius)
print(f"반지름 {circle_radius}인 원의 넓이: {circle_result:.2f}")

# 직육면체 표면적 및 부피 계산
length = 5
width = 3
height = 4
surface_area = mymath.rectangular_prism_area(length, width, height)
volume = mymath.rectangular_prism_volume(length, width, height)
print(f"가로 {length}, 세로 {width}, 높이 {height}인 직육면체의 표면적: {surface_area}")
print(f"가로 {length}, 세로 {width}, 높이 {height}인 직육면체의 부피: {volume}")
````
